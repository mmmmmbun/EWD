# Regresja liniowa

```python
# import bibliotek~
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generowanie danych

rng = np.random.RandomState(1)
x = 10 * rng.rand(30)
y = 3 * x - 3 + rng.randn(30)
plt.scatter(x, y);
```

Scikit-learn

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y,  label='original data')
plt.plot(xfit, yfit, "r", label='fitted line')
plt.legend()
```

SciPy

```python
from scipy import stats
res = stats.linregress(x, y)

print(f"R-squared: {res.rvalue**2:.6f}") #idealnie jeśli =1, rośnie jesi dodajemu wspolczynniki nawet jesli model nie jest lepszy (wtedy lepiej)
print(res)

plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
```

statsmodel

```python
import pandas as pd , numpy as np
from sklearn import datasets, linear_model
# from sklearn.linear_model import LinearRegression
import statsmodels.api as sma
```

```python
# X = data['TV']
# y = data['sales']
x2 = sma.add_constant(x)
model = sma.OLS(y, x2)
res = model.fit()
print(res.summary())

plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.fittedvalues, 'r', label='fitted line')
plt.legend()
plt.show()
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
cols =["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"]
boston = pd.DataFrame(data=data, columns=cols)
boston.isna().any()
Y = pd.DataFrame(target, columns=["PRICE"])
all = pd.concat([boston, Y], axis=1, join="inner")
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(all, test_size=0.20, shuffle=False)
```

```python
plt.hist(target)
plt.title("Mediana cen domów na własność (w tys. $)")
plt.show()
```

```python
train.corr()
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(10,10))   
sns.heatmap(train.corr(), annot=True)
```

```python
import statsmodels.formula.api as sm
model = sm.ols(data=train, formula='PRICE ~ CRIM+INDUS+CHAS+NOX+RM+DIS+RAD+PTRATIO+B+LSTAT+AGE+TAX+ZN').fit()
model.summary()
```

```python
model2 = sm.ols(data=train_smol, formula='PRICE ~ CRIM+INDUS+RM+PTRATIO+LSTAT+TAX').fit()
model2.summary()

# Spadło R-squared, Ale model jest dalej istotny statystycznie, ma tez wyzsze kryteria
```

```python
test_smol = test.drop(['ZN', 'CHAS', 'DIS', 'B', 'RAD','AGE','NOX'], axis=1)
```

```python
model3 = sm.ols(data=test_smol, formula='PRICE ~ CRIM+INDUS+RM+PTRATIO+LSTAT+TAX').fit()
model3.summary()
```

```python
predictions = model3.predict()
```

```python
from sklearn.metrics import mean_squared_error
print('MSE : ', mean_squared_error(test_smol["PRICE"],predictions))
print('RMSE : ', np.sqrt(mean_squared_error(test_smol["PRICE"],predictions)))
```

```python
# matlab prostszy bardziej intuicyjny szczegolnie dlauzytkownikow matlaba
plt.figure()  # utworzenie obrazu wykresu

# utworzenie pierwszego panelu i ustawienie osi
plt.subplot(2, 1, 1) # (liczba wierszy, liczba kolumn, numer panelu)
plt.plot(x, np.sin(x))

# drugi panel
plt.subplot(2,1, 2)
plt.plot(x, np.cos(x));
```

```python
# matlab prostszy bardziej intuicyjny szczegolnie dlauzytkownikow matlaba
plt.figure()  # utworzenie obrazu wykresu

# utworzenie pierwszego panelu i ustawienie osi
plt.subplot(2, 2, 1) # (liczba wierszy, liczba kolumn, numer panelu)
plt.plot(x, np.sin(x))

# drugi panel
plt.subplot(2,2, 4)  
plt.plot(x, np.cos(x));

#numerowanie obrazow od lewej do prawej góra dół
```

```python
# bardziej skomplikowany styl obiektowy, niezbedny do bardziej skomplikowanych wykresow

# Utworzenie obrazu fig oraz ax tablicy zawierającej 2 obiekty osi
fig, ax = plt.subplots(2)

# wywołananie plot na obydwu osiach
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python

# Utworzenie obrazu fig oraz ax tablicy dwuwymiarowej zawierającej 4 obiekty osi
fig, ax = plt.subplots(2,2)

# wywołananie plot na 2 osiach
ax[0][0].plot(x, np.sin(x))
ax[1][1].plot(x, np.cos(x));
```

```python
# mozna też użyć interfejsu do utworzenia prostego wykresu (jak w pierwszysm przykładzie)
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

# Rodzaje linii

```python
#różne style  linii
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# to samo w krótszzej formie 
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x - 0, color='blue')        # specify color by name
plt.plot(x, x - 1, color='g')           # short color code (rgbcmyk)
plt.plot(x, x - 2, color='0.75')        # Grayscale between 0 and 1
plt.plot(x, x - 3, color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, x - 4, color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, x - 5, color='chartreuse'); # all HTML color names supported
```

# Etykiety

```python

plt.plot(x, x)
plt.title("Tytuł wykresu")
plt.xlabel("x")
plt.ylabel("f(x)");
```

```python
plt.plot(x, 1/(x+0.1), '-g', label='1/(x+0.1)')
plt.plot(x, x, ':b', label='x')
plt.xlim(-1, 10)
plt.ylim(0, 15);
plt.legend();
```

```python
# Można ustwić wszystko w metodzie set

ax = plt.axes()
ax.plot(x, x)
ax.set(xlim=(0, 10), ylim=(0, 20),
       xlabel='x', ylabel='f(x)',
       title='A Simple Plot');
```

# Wykresy punktowe

```python
x = np.linspace(0, 10, 30)
y = x+2

plt.plot(x, y, "o", color='green'); # dodajemy "o" aby otrzymać punkty
```

```python
#inne parametry wykresu

x = np.linspace(0, 10, 4)
y = x+2


plt.plot(x, y, '-d', color='lightblue',
         markersize=10, linewidth=4,
         markerfacecolor='orange',
         markeredgecolor='green',
         markeredgewidth=2)
plt.ylim(0, 16);
```

```python
#możemy też użyć scatter, który daje więcej możliwosci wizualnych ale dla dużych danych jest wolniejszy

plt.scatter(x, y, marker='o');
```

```python
rnd = np.random.RandomState(0)

colors = rnd.rand(4)
sizes = [10, 20, 30, 60]

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  
```

# Wykres słupkowy

```python
import  numpy as np
values = np.array([990,1000,990,998,1005,986,987])   
idx = np.array(range(1,8))
ax = plt.axes()  
ax.set_title("Ciśnienie atmosferyczne w ciągu ostatnich 7 dni")
ax.bar(x=idx, height=values, color="lightblue") 
```

```python
#nieco czytelniej
values = np.array([990,1000,990,998,1005,986,987])   
idx = np.array(range(1,8))
ax = plt.axes()
ax.bar(x=idx, height=values, color="lightblue") 
ax.set(ylim=(920, 1050), title="Ciśnienie atmosferyczne w ciągu ostatnich 7 dni")
```

# Wykres pudelkowy

```python
wzrost_człowieka_grupaA =[177, 170, 166, 180,181, 188, 190,190,194, 210]
ax = plt.axes()

ax.boxplot(wzrost_człowieka_grupaA )

ax.set(ylim=(150, 230), title='Wzrost człowieka w pewnej grupie')
```

```python
wzrost_A =[177, 170, 166, 180,181, 188, 190,190,194, 210]
wzrost_B =[170, 170, 166, 160,151, 188, 171,160,154, 110]

data=[wzrost_A, wzrost_B]
ax = plt.axes()

ax.boxplot(data)
# ax.boxplot(wzrost_człowieka_grupaB )


ax.set(ylim=(100, 230), title='Wzrost człowieka w grupie A i B', xticklabels=["grupa A", "grupa B"])
```

```python
wzrost_A =[177, 170, 166, 180,181, 188, 190,190,194, 210]
wzrost_B =[170, 170, 166, 160,151, 188, 171,160,154, 110]

fig, ax = plt.subplots(1,2)
fig.suptitle('Wzrost człowieka w grupie A i B')

ax[0].boxplot(wzrost_A)
ax[1].boxplot(wzrost_B)

# ax.boxplot(wzrost_człowieka_grupaB )

ax[0].set(ylim=(100, 230),  xticklabels=["grupa A"]);
ax[1].set(ylim=(100, 230), xticklabels=["grupa B"]);
```
